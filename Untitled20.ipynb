```python
# A bank issues credit card to its customers under the scheme of Master card. Based on past data, it is known that 
# 60% of the customers pay the bill on time. If a sample of 7 accounts is selected at random from current database, 
# construct the binomial probability distribution of accounts paying on time.
```

```python
p=.6
n=7
```

```python
n=np.arange(0,8)
```

```python
n
```

```python
binomial = stats.binom.pmf(n,7,p)
```

```python
binomial
```

```python
binomial[2]
```

```python
plt.plot(n,binomial,'o-')
plt.xlabel('outcomes')
plt.ylabel('probabilities')
plt.show()
```
